```python
import pandas as pd
import numpy as np
```

# Clean main data

```python
df = pd.read_csv("cp.csv")
```

```python
df.head()
```

```python
# Drop Unnamed column (it's a copy of the index)
df.drop("Unnamed: 0", axis=1,inplace=True)
```

```python
# Delete columns that I don't understand (this is kinda sad hehe)
df.drop(["d_CP", "c_oficina", "c_CP", "id_asenta_cpcons", "c_cve_ciudad"],
       axis =1, inplace=True)
```

```python
df.head()
```

```python
# Remove accents and weird characters
cols = df.select_dtypes(include=[np.object]).columns
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
df.head()
```

```python
df.to_csv("cp_cleaned.csv", index=False)
```

# Clean lat lon data

```python
df_lat_lon = pd.read_csv("mx_postal_codes.csv")
```

```python
df_lat_lon.drop(["Unnamed: 6", "Unnamed: 7", "Unnamed: 8", "Unnamed: 9",
               "Unnamed: 10"], axis=1, inplace=True)
```

```python
df_lat_lon.head()
```

```python
df_lat_lon.dtypes
```

```python
df_lat_lon["Latitude"] = pd.to_numeric(df_lat_lon['Latitude'], errors='coerce')
df_lat_lon["Longtiude"] = pd.to_numeric(df_lat_lon['Longtiude'], errors='coerce')
```

```python
df_lat_lon.dtypes
```

```python
df_lat_lon.to_csv("mx_lat_lon.csv", index=False)
```
