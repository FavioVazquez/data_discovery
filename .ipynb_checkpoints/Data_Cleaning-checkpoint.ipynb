```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cp.csv")
```

```python
df.head()
```

```python
# Drop Unnamed column (it's a copy of the index)
df.drop("Unnamed: 0", axis=1,inplace=True)
```

```python
# Delete columns that I don't understand (this is kinda sad hehe)
df.drop(["d_CP", "c_oficina", "c_CP", "id_asenta_cpcons", "c_cve_ciudad"],
       axis =1, inplace=True)
```

```python
df.head()
```

```python
# Remove accents and weird characters
cols = df.select_dtypes(include=[np.object]).columns
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
df.head()
```

```python
df.to_csv("cp_cleaned.csv", index=False)
```
